## Amenities String 에서 나누는 방법

```python
import json
```

```python
full_list = []
```

```python
pattern = r'\"(.*?)\"'
pattern2 = r'\'(.*?)\''
nan = float('nan')
count = 0
for list_of_strings in Amenities:
  # print(list_of_strings)
  if isinstance(list_of_strings, float):
    count += 1
    continue
  else:
    matching_elements = re.findall(pattern, list_of_strings)
    if "{" in list_of_strings:
      element_dict = json.loads(list_of_strings)
      elements = []
      for key, value in element_dict.items():
        if value == "True" :
          elements.append(key)
        else :
          elements.append(value)
      elements = str(elements)
      elements = re.findall(pattern2, elements)
      matching_elements = elements
  for element_matched in matching_elements:
    full_list.append(element_matched)
  count +=1
```

```python
len(full_list)
```

```python
full_dataframe = pd.DataFrame({'items' : full_list})
unique_amenities = full_dataframe['items'].unique()
data_frame = pd.DataFrame(columns = unique_amenities)
data_frame
```
