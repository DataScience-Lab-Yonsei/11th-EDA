```python
def extract_policy(s):
  if isinstance(s, str):
      if 'strict' in s:
          return 'strict'
      elif 'moderate' in s:
          return 'moderate'
      elif 'flexible' in s:
          return 'flexible'
      else:
          return s
  else :
    return s
```

```python
canncellation_policy = cancellation_policy.astype(str)
canncellation_policy = cancellation_policy.apply(extract_policy)
```

```python
new_cancellation_policy = cancellation_policy.apply(extract_policy)
```
