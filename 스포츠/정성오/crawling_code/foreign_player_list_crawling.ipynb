## Change team name

### remove '[]' values like in 'kt wiz[B]'

```python
for column in foreign_player_2014_2024.columns:
    foreign_player_2014_2024[column] = foreign_player_2014_2024[column].astype('str')
```

```python
def remove_brackets(text):
    return re.sub(r"\[.*\]", "", text)

# Assuming foreign_player_2014_2024 is your DataFrame
foreign_player_2014_2024 = foreign_player_2014_2024.map(remove_brackets)
```

```python
foreign_player_2014_2024[foreign_player_2014_2024['year']=='2016']
```

### Capialization
ex. kt -> KT

```python
# Function to process the text
def process_text(text):
    # Extract the first two characters
    extracted = text.split()[0]
    # Check if the extracted text is in English and convert to uppercase
    return extracted.upper() if extracted.isalpha() and extracted.isascii() else extracted

# Apply the function to the '구단명' column
foreign_player_2014_2024['구단명'] = foreign_player_2014_2024['구단명'].apply(process_text)
```

## 선수명을 등록명으로 바꾸기

```python
pitcher_data = pd.read_csv('/Users/sjung/Desktop/Data_Science_Lab/EDA_project/datasets/pitcherRecords.csv')
hitter_data = pd.read_csv('/Users/sjung/Desktop/Data_Science_Lab/EDA_project/datasets/hitterRecords.csv')
pitcher_detailed_data = pd.read_csv("/Users/sjung/Desktop/Data_Science_Lab/EDA_project/datasets/pitcher_detailed_stats.csv")
hitter_detailed_data = pd.read_csv("/Users/sjung/Desktop/Data_Science_Lab/EDA_project/datasets/hitter_detailed_stats.csv")
```

```python
pitcher_data.head()
```

```python
pitcher_detailed_data.head()
```

```python
def find_name(names, team, position):
    if position == '투수':
        for name in names:
            enrolled_name = pitcher_data[(pitcher_data['이름']==name) & (pitcher_data['팀명']==team)]
            if not enrolled_name.empty:
                return enrolled_name['이름'].values[0]
            else:
                enrolled_name = pitcher_detailed_data[(pitcher_detailed_data['name']==name) &\
                                                        (pitcher_detailed_data['팀']==team)]
                if not enrolled_name.empty:
                    return enrolled_name['name'].values[0]
                
    else:
        for name in names:
            enrolled_name = hitter_data[(hitter_data['이름']==name) & (hitter_data['팀명']==team)]
            if not enrolled_name.empty:
                return enrolled_name['이름'].values[0]
            else:
                enrolled_name = hitter_detailed_data[(hitter_detailed_data['name']==name) &\
                                                        (hitter_detailed_data['팀']==team)]
                if not enrolled_name.empty:
                    return enrolled_name['name'].values[0]
            
    return "unknown"
    
```

```python
enrolled_name_list = []

for i in range(len(foreign_player_2014_2024)):
    names = foreign_player_2014_2024.loc[i, "성명"].split()
    position = foreign_player_2014_2024.loc[i, "포지션"]
    if len(names) > 1:
        for idx in range(len(names)-1):
            names.append("".join([names[idx], names[idx+1]]))
    team = foreign_player_2014_2024.loc[i, "구단명"]
    enrolled_name = find_name(names, team, position)
    enrolled_name_list.append(enrolled_name)
```

```python
# index of players who don't have any records in Korea
np.where(np.array(enrolled_name_list) == "unknown")
```

```python
foreign_player_2014_2024.insert(2, "등록명", enrolled_name_list)
```
