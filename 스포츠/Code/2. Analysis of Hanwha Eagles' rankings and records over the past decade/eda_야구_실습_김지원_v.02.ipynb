```python
# 먼저 'foreign_player_list.csv' 에서 '등록명'을 'name'으로 변경합니다.
foreign_player_list.rename(columns={'등록명': 'name'}, inplace=True)
```

```python
# 데이터 병합
merged_data = foreign_player_list.merge(pitcher_detailed_stats[['name', 'WAR']], on='name', how='inner')

# 합쳐진 데이터의 결과를 반환
merged_data.head()
```

```python
# SK-->SSG,넥센-->키움
merged_data['구단명'] = merged_data['구단명'].replace({'SK': 'SSG', '넥센': '키움'})
```

```python
merged_data['구단명'].unique()
```

```python
# 2014-2023년 사이의 데이터만 필터링
filtered_data = merged_data[(merged_data['year'] >= 2014) & (merged_data['year'] <= 2023)]

# 각 팀별로 외국인 선수들의 WAR 값을 합산
team_foreign_war = filtered_data.groupby(['구단명', 'year'])['WAR'].sum().reset_index()


# Plotly로 시각화를 진행합니다.
fig = px.scatter(team_foreign_war, x='year', y='WAR', color='구단명', color_discrete_map=team_colors,
                 title='2014-2023년 각 팀 외국인 선수 WAR 비교')

# 한화 팀에 대해서만 선으로 연결합니다.
hanwha_foreign_war = team_foreign_war[team_foreign_war['구단명'] == '한화']
fig.add_trace(px.line(hanwha_foreign_war, x='year', y='WAR', color='구단명', color_discrete_map=team_colors).data[0])

# 제목을 중앙으로 정렬하고 x축의 모든 연도를 표시합니다.
fig.update_layout(
    title_text="2014-2023년 각 팀 외국인 선수 WAR 비교",
    title_x=0.5,
    xaxis=dict(
        tickmode='linear',
        dtick=1,
        range=[2013.5, 2023.5] # 연도가 소수점으로 표현되지 않도록 범위를 설정합니다.
    )
)


fig.show()
```

# 분석 포인트 5: 각 팀 10년 간 (2014-2023년) 전체 승수에서 외국인 선수들의 WAR 비교


```python
pitcher_detailed_stats_re = pd.read_csv('/content/drive/MyDrive/03)DSL/2_EDA_프로젝트/pitcher_detailed_stats_revised_1.csv')
```

```python
# 팀별 전체 승수 계산
pitcher_detailed_stats_filtered= pitcher_detailed_stats_re[(pitcher_detailed_stats_re['연도'] >= 2014) & (pitcher_detailed_stats_re['연도'] <= 2023)]
team_wins = pitcher_detailed_stats_filtered.groupby('팀')['승'].sum().reset_index()

# 필요한 팀 목록을 정의합니다.
teams = ['KIA', 'KT', 'LG', 'NC', 'SSG', '키움', '두산', '롯데', '삼성', '한화']

# '팀' 컬럼에서 위에 정의된 팀들만 필터링합니다.
team_wins_filtered = team_wins[team_wins['팀'].isin(teams)]

team_wins_filtered = team_wins[~team_wins['팀'].str.contains(",")]
team_wins_filtered.head()

# 데이터 병합
team_war_contribution = team_wins_filtered.merge(team_foreign_war, left_on='팀', right_on='구단명', how='left')

# WAR 비중 계산
team_war_contribution['WAR'] = team_war_contribution['WAR'].fillna(0)
team_war_contribution['외국인선수_WAR_비중'] = team_war_contribution['WAR'] / team_war_contribution['승']
team_war_contribution.head()
```

```python
team_war_contribution['구단명'].unique()
```

```python
# 순위 계산
#team_war_contribution['외국인선수_WAR_비중_순위'] = team_war_contribution.groupby('year')['외국인선수_WAR_비중'].rank(ascending=False)

# 순위 계산 시 'method'를 'first'로 설정하고, 결과를 'Int64'로 변환하여 NA 값을 허용하는 정수형으로 만듭니다.
team_war_contribution['외국인선수_WAR_비중_순위'] = (
    team_war_contribution
    .groupby('year')['외국인선수_WAR_비중']
    .rank(ascending=False, method='first')
    .astype('Int64')  # NA 값이 포함된 정수형으로 변환
)

# Plotly로 시각화를 진행합니다.
fig = px.scatter(team_war_contribution, x='year', y='외국인선수_WAR_비중', color='구단명', color_discrete_map=team_colors,
                 title='팀별 전체 승수 대비 외국인선수 WAR 비중')

# 한화 팀에 대해서만 선으로 연결합니다.
hanwha_foreign_percentage = team_war_contribution[team_war_contribution['구단명'] == '한화']
fig.add_trace(px.line(hanwha_foreign_percentage, x='year', y='외국인선수_WAR_비중', color='구단명', color_discrete_map=team_colors).data[0])

# 한화 팀의 각 데이터 포인트에 순위 표시
for idx, row in hanwha_foreign_percentage.iterrows():
    fig.add_annotation(x=row['year'], y=row['외국인선수_WAR_비중'], text=str(row['외국인선수_WAR_비중_순위']) + '위', showarrow=True, arrowhead=1)


# 제목을 중앙으로 정렬하고 x축의 모든 연도를 표시합니다.
fig.update_layout(
    title_text="팀별 전체 승수 대비 외국인선수 WAR 비중",
    title_x=0.5,
    xaxis=dict(
        tickmode='linear',
        dtick=1,
        range=[2013.5, 2023.5] # 연도가 소수점으로 표현되지 않도록 범위를 설정합니다.
    )
)

# 시각화

fig.show()
```
