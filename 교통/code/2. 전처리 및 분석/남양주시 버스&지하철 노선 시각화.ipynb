```python
import pandas as pd
bus_list=['11','100','1000','1000-1','1001','1003','1006','1100','1200','1200-1','1660','1670','1670-1','1700','2000','2000-1','7007','8001','8002','8002-1','8012','G1200','G1300','G1300N','G6000','G6000N','G6100','G6100N','G9311','M2316','M2323','M2341','M2344','M2352','M2353'] # 버스 노선 입력
df_bus = pd.DataFrame()
for i in bus_list:
  bus = '남양주_{}'.format(i)
  for j in range(1,5): # 1~4까지 - 데이터명: 지역_노선_번호 일때 가정
    filepath = '/content/drive/MyDrive/Colab Notebooks/DSL/EDA_Project/남양주버스/'
    filename = bus+'_{}.xlsx'.format(j)
    df = pd.read_excel(filepath+filename)
    df = df.drop([0],axis=0)
    df = df.drop(['기종점','정류장순번'],axis=1)
    for l in range(1,len(df)+1):
      fill_word = df['노선'][1]
      df.at[l,'노선'] = fill_word

    for l in range(1,len(df)+1):
      fill_word = df['월'][1]
      df.at[l,'월'] = fill_word

    for l in range(1,len(df)+1):
      if df['정류장'][l] != 0:
        fill_word = df['정류장'][l]
      else:
        df.at[l,'정류장'] = fill_word

    df_bus = pd.concat([df_bus, df],ignore_index=True)
    df_bus = df_bus[(df_bus.이용자유형=='일반인')]
df_bus.head()
```

```python
import numpy as np
import pandas as pd
import json
import requests
import branca
import folium
from branca.colormap import linear

filepath = '/content/drive/MyDrive/Colab Notebooks/DSL/EDA_Project/'
filename = '기업수.xlsx'
seoul_data = pd.read_excel(filepath+filename)
colormap = linear.Reds_04.scale(
    np.min(seoul_data['기업']),
    np.max(seoul_data['기업'])
)
colormap
```

```python
population_dict = seoul_data.set_index('지역')['기업']
color_dict = {str(key): colormap(population_dict[key]) for key in population_dict.keys()}
color_dict
```
