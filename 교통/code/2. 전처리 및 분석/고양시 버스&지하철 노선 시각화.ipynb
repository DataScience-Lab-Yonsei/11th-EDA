```python
import pandas as pd
bus_list=['1000','1001','1082','1100','1200','1500','1900','200','9600','9700','9701','9707','9711','9714','M7119','M7129','M7412','M7646','M7731'] # 버스 노선 입력
df_bus = pd.DataFrame()
for i in bus_list:
  bus = '고양_{}'.format(i)
  for j in range(1,5): # 1~4까지 - 데이터명: 지역_노선_번호 일때 가정
    filepath = '/content/drive/MyDrive/Colab Notebooks/DSL/EDA_Project/고양시버스/'
    filename = bus+'_{}.xlsx'.format(j)
    df = pd.read_excel(filepath+filename)
    df = df.drop([0],axis=0)
    df = df.drop(['기종점','정류장순번'],axis=1)
    for l in range(1,len(df)+1):
      fill_word = df['노선'][1]
      df.at[l,'노선'] = fill_word

    for l in range(1,len(df)+1):
      fill_word = df['월'][1]
      df.at[l,'월'] = fill_word

    for l in range(1,len(df)+1):
      if df['정류장'][l] != 0:
        fill_word = df['정류장'][l]
      else:
        df.at[l,'정류장'] = fill_word

    df_bus = pd.concat([df_bus, df],ignore_index=True)
    df_bus = df_bus[(df_bus.이용자유형=='일반인')]
df_bus.head()
```
